

### Dataset Summary:

This dataset contains traffic accident data in Pakistan, with information collected over multiple years. It provides details on the total number of accidents, fatalities, and non-fatal accidents, categorized by year and city. The data can be used to analyze trends in road accidents, including the number of accidents, fatalities, and the locations most affected.

##### Columns in the dataset:

1. Year: The year in which the traffic accidents occurred.


2. City: The city in which the traffic accidents occurred.


3. Total number of accidents: The total number of traffic accidents that occurred in each city for each year.


4. Killed: The total number of people killed in traffic accidents in each city for each year.


5. Non-Fatal Accidents: The total number of accidents that did not result in fatalities in each city for each year.


6. Fatal Accident: The total number of fatal accidents in each city for each year.



##### Purpose of the Dataset:

The dataset can be used to understand:

The distribution of traffic accidents across different years and cities in Pakistan.

The correlation between accident rates and fatalities.

Trends in traffic safety and accident prevention over time.


This dataset provides valuable insights into road safety, and it can be used to guide policy-making, traffic management strategies, and public awareness campaigns to reduce traffic-related incidents.

```python


# Import plotly.express and pandas
import plotly.express as px
import pandas as pd

# Sample data 
data = pd.DataFrame({
    'Place': ['Sindh', 'Punjab', 'Balochistan', 'Khyber Pakhtunkhwa', 'Sindh', 'Punjab'],
    'Killed': [10, 20, 15, 5, 8, 12]
})

# List of places to filter
places = ['Sindh', 'Punjab', 'Balochistan', 'Khyber Pakhtunkhwa']

# Filter the data based on the places
selected_data = data.loc[data['Place'].isin(places), ['Place', 'Killed']]

# Create the pie chart
fig = px.pie(
    selected_data,
    values='Killed',
    names='Place',
    width=950,
    height=900,
    color='Place',
    color_discrete_map={
        'Punjab': 'lightcyan',
        'Sindh': 'cyan',
        'Balochistan': 'royalblue',
        'Khyber Pakhtunkhwa': 'darkblue'
    },
    title='Ratio of Deaths Due to Accidents in Provinces'
)

# Display the chart
fig.show()

```

```python

# Import plotly.express and pandas
import plotly.express as px
import pandas as pd

# Sample data 
data = pd.DataFrame({
    'Place': ['Islamabad', 'Islamabad', 'Islamabad', 'Karachi', 'Islamabad'],
    'Year': [2018, 2019, 2020, 2020, 2021],
    'Total_accidents': [50, 70, 80, 60, 90],
    'Killed': [5, 7, 10, 4, 8]
})

# Filter the data for total accidents in Islamabad
sel = data[['Year', 'Total_accidents']].loc[(data['Place'] == 'Islamabad')]

# Create scatter plot for total accidents
graph = px.scatter(
    sel,
    x='Year',
    y='Total_accidents',
    color='Year',
    size='Total_accidents',
    title='Total Accidents in Islamabad'
)

# Show the first graph
graph.show()

# Filter the data for total killed in Islamabad
sel1 = data[['Year', 'Killed']].loc[(data['Place'] == 'Islamabad')]

# Create scatter plot for total killed
graph1 = px.scatter(
    sel1,
    x='Year',
    y='Killed',
    color='Year',
    size='Killed',
    title='Total People Killed in Islamabad Through Traffic Accidents'
)

# Show the second graph
graph1.show()

```

```python

# Import necessary libraries
import plotly.express as px
import pandas as pd

# Sample data (replace this with your actual data if different)
data = pd.DataFrame({
    'Place': ['Pakistan', 'Pakistan', 'Pakistan', 'Pakistan', 'Pakistan', 'India', 'Pakistan'],
    'Year': [2018, 2019, 2020, 2021, 2022, 2020, 2023],
    'Non-Fatal_Accidents': [500, 650, 700, 800, 750, 600, 820]
})

# Display the sample data
print("Sample Data:")
print(data)

# Filter data for Non-Fatal Accidents in Pakistan
sel = data[['Year', 'Non-Fatal_Accidents', 'Place']].loc[data['Place'] == 'Pakistan']

# Display the filtered data
print("\nFiltered Data for Pakistan:")
print(sel)

# Create bar chart for Non-Fatal Accidents in Pakistan
graph = px.bar(
    sel,
    x='Year',
    y='Non-Fatal_Accidents',
    color='Year',
    text='Non-Fatal_Accidents',
    title='Total Non-Fatal Accidents in Pakistan'
)

# Update text position to display outside the bars
graph.update_traces(texttemplate='%{text}', textposition='outside')

# Update layout for better visuals (optional)
graph.update_layout(
    uniformtext_minsize=8,
    uniformtext_mode='hide',
    xaxis=dict(tickmode='linear'),
    yaxis_title='Number of Non-Fatal Accidents',
    xaxis_title='Year',
    legend_title='Year'
)

# Show the bar chart
graph.show()

```
